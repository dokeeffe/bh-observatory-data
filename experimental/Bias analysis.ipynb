```python
import os
import numpy as np
from astropy import units as u
from ccdproc import CCDData
from scipy import stats
import matplotlib.pyplot as plt
from photutils import aperture_photometry, CircularAperture
%matplotlib inline
plt.style.use('classic')

bias = CCDData.read('Bias__001.fits', unit=u.adu)
master_bias = CCDData.read('MasterBias/master_bias-25_2X.fits', unit=u.adu)
```

```python
print('min {}, max {}, avg {}, std {},'.format(bias.data.min(),bias.data.max(),bias.data.mean(),bias.data.std()))
print('min {}, max {}, avg {}, std {},'.format(master_bias.data.min(),master_bias.data.max(),master_bias.data.mean(),master_bias.data.std()))

```

```python
fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
plt.imshow(bias.data, cmap='gray', origin='lower', vmin=350, vmax=600)
```
